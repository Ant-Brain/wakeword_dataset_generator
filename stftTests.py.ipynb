```python
import librosa
import librosa.display # for waveplots, spectograms, etc
import IPython.display as ipd # for playing files within pythonimport numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x , sample_rate = librosa.load("dataset_format_fixed/achieve/Tessa_SIRI.wav",sr=16000)
print(len(x),sample_rate)

number_of_bits = 16000-len(x) 

```

```python
ipd.Audio(data=x*2**15,rate=sample_rate)

import numpy as np
import random 

frontBits = random.randint(0,number_of_bits)

plt.figure(figsize=(14,5))

new_x =np.append(np.zeros(frontBits),x)
new_x = np.append(new_x,np.zeros(number_of_bits-frontBits))
librosa.display.waveplot(new_x*0.6+new_x*0.4,sample_rate)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(x,sample_rate)
```

```python
n_fft, hop_length, n_mels, sr = 512,512, 128, 16000

#%%
mel_spectrogram = librosa.feature.melspectrogram(
    x, sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
librosa.display.specshow(log_mel_spectrogram, x_axis="time", y_axis="mel", sr=sr)
print(log_mel_spectrogram.shape)
plt.show()
```
